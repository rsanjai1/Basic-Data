```python
d2 = d1.fillna(0)
```

```python
x = d2[['ApplicantIncome','CoapplicantIncome','LoanAmount',
       'Loan_Amount_Term','Credit_History']]
y = d2['Loan_Status']
```

```python
g1 = {"Loan_Status":{'Y':1,'N':2}}
d2= d2.replace(g1)
print(d2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.70)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
y_train.shape
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
parameters = {
    'max_depth':[10,20,50,60,30],
    'min_samples_leaf':[35,16,75,70,45],
    'n_estimators':[10,20,30,40,50]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator=rfc,param_grid=parameters,cv=2,scoring='accuracy')
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_score_
```

```python
from sklearn.tree import plot_tree
```

```python
rfc_best= grid_search.best_estimator_
```

```python
plt.figure(figsize=(80,60))
plot_tree(rfc_best.estimators_[5],feature_names=x.columns,class_names=["Yes","No"],filled=True)
```
